```python
import csv
import numpy

with open('/content/values.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)

    line_count = 0
    map = {}

    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            gender = row[3]
            hispanic = row[5]
            race = row[6]
            sexual_orientation = row[2]
            feats = [gender, hispanic, race, sexual_orientation]
            permutation = ','.join(feats)

            if permutation not in map:
              map[permutation] = {
                  "count": 0,
                  # hispanic
                  "Hispanic": 0,
                  "Non-Hispanic": 0,
                  # race
                  "White": 0,
                  "American Indian, Aleut, or Eskimo": 0,
                  "Asian or Pacific Islander": 0,
                  "Black or African American": 0,
                  "Other": 0,
                  # other
                  "agewhenmet": [],
                  "timefrommettorel": [],
                  "agewhenmarry": [],
              }

            p_hispanic = row[7]
            p_race = row[8]
            age_when_met = int(row[9])
            time_from_met_to_rel = float(row[10])
            age_when_marry = int(row[11])
            map[permutation]["count"] += 1
            map[permutation][p_hispanic] += 1
            map[permutation][p_race] += 1
            map[permutation]["agewhenmet"].append(age_when_met)
            map[permutation]["timefrommettorel"].append(time_from_met_to_rel)
            map[permutation]["agewhenmarry"].append(age_when_marry)

        line_count += 1

    maps = []
    for k in map:
      items = k.split(',')
      gender = items[0]
      hispanic = items[1]
      race = items[2]
      sexual_orientation = items[3]
      map_final = { "permutation": k, "gender": gender, "hispanic": hispanic, "race": race, "sexual_orientation": sexual_orientation, **map[k] }
      maps.append(map_final)
    print(maps)
    print(len(maps))

    field_names = ["permutation",
                  "gender",
                  "hispanic",
                  "race",
                  "sexual_orientation",
                  # from map
                  "count",
                  # hispanic
                  "Hispanic",
                  "Non-Hispanic",
                  # race
                  "White",
                  "American Indian, Aleut, or Eskimo",
                  "Asian or Pacific Islander",
                  "Black or African American",
                  "Other",
                  # other
                  "agewhenmet",
                  "timefrommettorel",
                  "agewhenmarry"]
    
    with open('outputs.csv', 'w') as csvfile:
      writer = csv.DictWriter(csvfile, fieldnames=field_names)
      writer.writeheader()
      writer.writerows(maps)

    print(f'Processed {line_count} lines.')
```
